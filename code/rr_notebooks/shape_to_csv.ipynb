## Converting `.shp` files to `.csv`

Our fire data was given to us as `.shp` files, which include geometry usable by programs like Tableau in addition to a csv-like structure with data. 

In order to pull that information out, we used geopandas to import the files and export them as `.csv` instead.

```python
import geopandas

geodf = geopandas.read_file('../../data/fire/mtbs_FODpoints_DD/mtbs_FODpoints_DD.shp')
```

```python
geodf_filtered = geodf.loc[geodf['Ig_Date'] >= '2009-12-31'].copy()
```

```python
geodf_filtered
```

```python
geodf_filtered.to_csv('gdf_filtered.csv')
```
